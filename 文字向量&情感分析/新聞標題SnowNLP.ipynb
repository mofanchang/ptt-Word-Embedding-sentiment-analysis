```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from bs4 import BeautifulSoup

def main():
    resp = requests.get('https://www.ettoday.net/news/focus/%E6%B0%91%E7%94%9F%E6%B6%88%E8%B2%BB/')
    soup = BeautifulSoup(resp.text, 'html.parser')

    # 取得第一篇 blog (h3)
    print(soup.find('h3'))
    print(soup.h3)  # 與上一行相等
      
    # 取得第一篇 blog 主標題(只取文字)
    print(soup.h3.text)    
    print(soup.h3.a.text)# 與上一行相等
    print(soup.h3.a.string)      

if __name__ == '__main__':
    main()
```

```python
import requests
from bs4 import BeautifulSoup

def main():
    resp = requests.get('https://www.ettoday.net/news/focus/%E6%B0%91%E7%94%9F%E6%B6%88%E8%B2%BB/')
    soup = BeautifulSoup(resp.text, 'html.parser')

    # 取得所有 blog 主標題, 使用 tag
    main_titles = soup.find_all('h3')
    for title in main_titles:
      print(title.a.text)

if __name__ == '__main__':
    main()
```

```python
# Open the original file for reading
with open("/content/drive/MyDrive/Colab Notebooks/lifeismoney-2.txt", "r") as original_file:
    # Read the contents of the file into memory
    contents = original_file.read()
    
    # Split the contents into a list of words
    words = contents.split()
    
    # Define the words to be removed
    words_to_remove = ['標題', '[情報]', ':', '[新聞]','Re']
    
    # Remove the unwanted words from the list of words
    filtered_words = [word for word in words if word.lower() not in words_to_remove]
    
    # Join the filtered words back into a single string
    filtered_contents = ' '.join(filtered_words)
    
# Open a new file for writing
with open("/content/drive/MyDrive/Colab Notebooks/lifeismoney-2-new.txt", "w") as filtered_file:
    # Write the filtered contents to the new file
    filtered_file.write(filtered_contents)
```
